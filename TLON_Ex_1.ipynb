```python
from tlon_util import tlon, view_tlon
import numpy as np
import matplotlib.pyplot as plt
```

```python
infile = '/home/nfeatherstone/surveys/modulated/work/HFJ_54_Pm_8/new_tlon_extra/q803_r0_50000000_60000000'
omega_sun = 2.87e-6
omega0=3.0*omega_sun   # rotation rate of this model

pi = np.pi
tau_omega = 2*pi/omega0  # rotation period in seconds
```

```python
bphi = tlon(infile)
tau = (bphi.time-bphi.time[0])/tau_omega
```

```python
# We can see what latitudes are contained
for i in range(bphi.nlat):
    print(i, 'lat = ', bphi.lat[i])
```

```python
lat_inds = [2,5,6]  # Pick a few latitudes to plot
```

```python
fig, ax = plt.subplots(figsize=(15,7),ncols=len(lat_inds))
for i, l in enumerate(lat_inds):
    
    lat_val = bphi.lat[l]
    lstring="{:.1f}".format(lat_val)
    tstring = 'latitude:  '+lstring+r'$^\circ$'
    
    #can set the time axis-limits
    ylim = [0,2000]
    
    view_tlon(bphi.vals[l,:,:],tau,bphi.lon,ax[i],scale_factor=1.5,
          color_map='RdYlBu_r',time_units=r'$\tau_\Omega$', 
          title = tstring, fig=fig, q_units='G', ylim = ylim)
plt.tight_layout()
```
